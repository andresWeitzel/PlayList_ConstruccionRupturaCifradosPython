```python
mensaje="mensajepararomper"

diccionario="abcdefghijklmnopqrstuvwxyz"

auxiliar=""

for i in mensaje:
    #La logica es similar al cifrado inverso, el punto aca es siempre a cada caracter del  
    #mensaje respecto a la posicion de ese caracter en el diccionario sumarle 13 posiciones, 
    #y teniendo en consideracion que no debe superar las 26 posiciones, osea todo el alfabeto,
    #por eso modulo 26, si llega al mismo, comienza desde el inicio.
    auxiliar = auxiliar + diccionario[(diccionario.find(i) + 13 ) % 26]
   
        
print("\nMensaje cifrado ROT13: "+auxiliar)    
```

```python
'''
Para el caso del cifrado cesar basico solamente sera posible usar el diccionario basico, ya que 
el algoritmo implementado limita a modulo 25, osea el caracter maximo de el abecedario y vamos a
obtener siempre el mensaje original, porque el algoritmo codificado es limitado, seguido de esto,
vamos a implementar el otro algoritmo mas extenso y veremos que los potenciales mensajes originales 
van a ser diferentes
'''


def rupturaMensaje(tipoDiccionario, mensaje):
    
    for key in range(len(tipoDiccionario)):
        
        auxiliar=""
        
        for i in mensaje:
    
            auxiliar = auxiliar + diccionario[(diccionario.find(i) + 13 ) % 26]
   
        
        print("\nKey numero : "+str(key)+" | Potencial Mensaje: "+auxiliar)  
        
```

```python
'''
Para este codigo, no se limita el modulo del diccionario, se trabaja en base a los indices e if's.
Notar que combinacionKEY la usamos para trbajar todas las posibles combinaciones en base a la longitud
del diccionario implementado y ademas por cada vuelta del primer for le restamos al caracter del mensaje
respecto del diccionario la longitud del diccionario.
'''
def rupturaMensaje02(tipoDiccionario, mensaje):
    
    for combinacionKEY in range(len(tipoDiccionario)):
        
        auxiliar=""
        
        
        for caracter in mensaje:

            if caracter in tipoDiccionario:

                 diccionarioIndice=tipoDiccionario.find(caracter)

                 auxiliarIndice=diccionarioIndice - combinacionKEY

                
                 if auxiliarIndice >= len(diccionario):

                     auxiliarIndice = auxiliarIndice - len(tipoDiccionario)


                 elif auxiliarIndice < 0:

                     auxiliarIndice = auxiliarIndice + len(tipoDiccionario)

    
    
                 auxiliar = auxiliar + tipoDiccionario[auxiliarIndice]


            else:
                 auxiliar= auxiliar + caracter



        print("\nKey numero : "+str(combinacionKEY)+" | Potencial Mensaje: "+auxiliar)  
        
```
